```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df.head()
```

```python
## To check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart
plt.pie(country_val,labels=country_names)
plt.show()
```

```python
## top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

Observation: zomato max records/transactions are from india followed by USA and then UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

Observation: 
1. when rating is between 4.5 and 4.9 -> Excellent
2. when rating is between 4 and 4.4 -> Very Good
3. when rating is between 3.5 and 3.9 -> Good
4. when rating is between 2.5 and 3.4 -> Average
5. when rating is between 1.8 and 2.3 -> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) # make figure bigger
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

Observation:
1. Not rated count is very high
2. Max number of ratings are between 2.5 and 3.4

```python
## count plot
sns.countplot(x="Rating color",data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

```python
### find countries' names giving zero rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

Observation:
Maximum of zero ratings are from india 

```python
### find which currency is used by which country
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
### which countries have online deliveries option
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts()
```

```python
final_df[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```

Observations:
1. Online deliveries are available in india and UAE
